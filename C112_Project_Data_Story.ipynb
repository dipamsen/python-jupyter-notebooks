# Project C112 - Data Story
Analysis and Visualization of Savings Dataset.

```
import pandas as ps
import statistics
import plotly.express as px
import plotly
plotly.io.templates.default = "plotly_dark"
import csv
import plotly.graph_objects as go
import plotly.figure_factory as ff

def get_stats(series):
  return ps.Series([series.mean(), series.median(), series.mode()[0], series.min(), series.max()], index=["Mean", "Median", "Mode", "Min", "Max"])
```

The Dataset:

```
df = ps.read_csv("https://raw.githubusercontent.com/whitehatjr/datasets/master/savings_data_final.csv")
df
```

## Scatter Plot
This graph is showing each entry as a point the quantity saved, and the age, and the color is determined by reminder given. We can see that the data is mostly scattered, and most people's savings amount lies in the range 0 - 1000.

```
px.scatter(df, x="age", y="quant_saved", color="rem_any")
```

## Measures of Central Tendency

```
get_stats(df["quant_saved"])
```

- The Mean amount saved is 85, even though the data ranges from 0 to 7353, which suggests that the data is concentrated at the bottom.
- Also, as the Mean differs from the Median, the data is not in a normal distribution

## Data Sampling

Select a Random set of 100 entries as a sample

```
sample1 = df["quant_saved"].sample(100)
sample2 = df["quant_saved"].sample(100)
```

```
get_stats(sample1)
```

```
get_stats(sample2)
```

## Standard Deviation

```
df['quant_saved'].std()
```

```
sample1.std()
```

```
sample2.std()
```

## Correlation
Correlation Coefficient for Age and Savings

```
df['quant_saved'].corr(df['age'])
```

The Age and Quantity saved are not corellated. The data is scattered over a wider range.

## Savings Distribution Graph

```
ff.create_distplot([df['quant_saved']], ["Quantity Saved"], show_hist=False)
```

The amount saved is highest in the range 0 - 1000, even though the max goes up to 7000
